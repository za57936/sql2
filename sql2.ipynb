```python
from sqlalchemy import create_engine     #importing create_engine
#create a engine to store data in memory
engine = create_engine('sqlite:///:memory:', echo=False)
from sqlalchemy.ext.declarative import declarative_base
import pandas as pd   #importing pandas library
Base = declarative_base() # constructing a base class
from sqlalchemy import Column, Integer, String
```

```python
class User(Base): #class definition that sql-chemy uses to create table
    __tablename__ = 'adult' # tablename adult
    
    id=Column(Integer, primary_key=True) #int id
    age =Column(Integer) #int age
    workclass=Column(String) #varchar workclass
    fnlwgt=Column(Integer) #integer fnlwgt
    education=Column(String) #varchar education
    education_num=Column(Integer) #integer eduction
    marital_status=Column(String) #varchar martial_status
    occupation=Column(String) #varchar occupation
    relationship=Column(String) #varchar relationship
    race=Column(String) #varchar race
    sex=Column(String) #varchar sex
    capital_gain=Column(Integer) #integer capital_gain
    capital_loss=Column(Integer) #integer capital_loss
    hours_per_week=Column(Integer) #integer hours_per_week
    native_country=Column(String) #varchar native_country
    classs=Column(String) #varchar classs
    
    def __repr__(self):
        return "<User(id='%d')>" % (self.id)
    
#Create all tables in the engine which is equivalent to "Create Table" statements in raw SQL.    
Base.metadata.create_all(engine)
```

```python
#2
# update-1
from sqlalchemy import update   #importing library
update(User).where(User.education=='Assoc-voc').values(native_country='Senior') #update table where education='Assoc-voc'
session.commit()


# update-2
update(User).where(User.race=='Amer-Indian-Eskimo').values(classs ='1111K') #update table where race=='Amer-Indian-Eskimo'
session.commit()     #committing
```

```python
#3
#deletion 1
session.query(User).filter(User.id == 10).delete() #deleting row where adult.id=10
session.commit()  #committing
print(session.query(User).all())  #printing outputs

# deletion-2
#deleting tuple where race='Asian-Pac-Islander' and occupation=='Craft-repair'
session.query(User).filter(User.race=='Asian-Pac-Islander').filter(User.occupation=='Craft-repair').delete()
session.commit()  # committing
print(session.query(User).all())  #printing outputs
```

```python
#4
#printing tuples where native_country=='United-States' and sex=='Male'
for user in session.query(User).filter(User.native_country=='United-States').filter(User.sex=='Male'):
    print(user)
# printing tuples where education like('%helors') order by Adult.id
query = session.query(User).filter(User.education.like('%helors')).order_by(User.id)
for item in query.all():
    print(item)
```
